```
import pandas as pd
import matplotlib.pyplot as plt

# --------------------------
# Step 1: Create a new dataset (3 inputs + 1 output)
# --------------------------
data = {
    "Square_Footage": [1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400,
                       1550, 1650, 1750, 1850, 1950, 2050, 2150, 2250, 2350, 2450],
    "Bedrooms":       [3, 3, 3, 4, 4, 4, 5, 5, 5, 6,
                       3, 4, 4, 4, 5, 5, 5, 6, 6, 6],
    "Age_of_House":   [5, 10, 8, 3, 7, 12, 15, 20, 1, 6,
                       4, 9, 11, 2, 8, 14, 18, 22, 3, 5],
    "Price":          [300000, 320000, 340000, 360000, 380000, 400000, 410000, 420000, 450000, 480000,
                       310000, 330000, 350000, 370000, 390000, 405000, 415000, 425000, 455000, 485000]
}

df = pd.DataFrame(data)
print("--- Dataset Head ---")
print(df.head())
print("-" * 20)

# --------------------------
# Step 2: Boxplots of input features
# --------------------------
# This creates three separate boxplots side-by-side
df[["Square_Footage","Bedrooms","Age_of_House"]].plot(kind="box",
                                       subplots=True, layout=(1,3),
                                       figsize=(15,5), patch_artist=True,
                                       flierprops=dict(markerfacecolor='red', marker='o'))
plt.suptitle("Boxplots of Input Features (Median, IQR, Outliers)", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --------------------------
# Step 3: Histograms of input features
# --------------------------
# This creates three separate histograms
df[["Square_Footage","Bedrooms","Age_of_House"]].hist(bins=8, figsize=(15,5), edgecolor="black", grid=False)
plt.suptitle("Histograms of Input Features with Distribution", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
